# Bài 1. Làm việc với file csv:
## 1. Đọc dữ liệu Covid được lưu ở file covid_19_clean_complete.csv (https://www.worldometers.info/)

```python
import pandas as pd
df = pd.read_csv('covid_19_clean_complete.csv')
df.head()
```

```python
df.info() #method prints information about a DataFrame including the index dtype and columns, non-null values and memory usage.
```

```python
df.shape #Return a tuple representing the dimensionality of the DataFrame.
```

```python
df.describe() #Generate descriptive statistics.
```

## 2. Thống kê số lượng nhiễm, từ vong, khỏi bệnh theo từng quốc gia.

**pandas.DataFrame.groupby** Group DataFrame using a mapper or by a Series of columns.

```python
df1 = df.groupby('Country/Region')[['Confirmed','Deaths','Recovered']].sum().sort_values(['Confirmed'], ascending=False)
df1
```

## 3. Thống kê số lượng nhiễm, tử vong, khỏi bệnh theo từng vùng (WHO region)


```python
df2 = df.groupby('WHO Region')[['Confirmed','Deaths','Recovered']].sum()
df2
```

## 4. Thống kê số lượng nhiễm, tử vong, khỏi bệnh theo từng ngày.


```python
df3 = df.groupby('Date')[['Confirmed','Deaths','Recovered']].sum()
df3
```

## 5. Lưu kết quả 2,3,4 vào một file results.xlsx với mỗi bảng thống kê ở một sheet khác nhau


**pandas.DataFrame.to_excel** Write object to an Excel sheet.

```python
with pd.ExcelWriter('results.xlsx') as writer:
    df1.to_excel(writer, sheet_name='Country')
    df2.to_excel(writer, sheet_name='WHO Region')
    df3.to_excel(writer, sheet_name='Date')
```
